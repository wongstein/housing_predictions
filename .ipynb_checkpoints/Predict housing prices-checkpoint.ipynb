Figure out what features we have, and normalize.  Log should be fine for now.

```python
target = 'median_house_value'
features_all = [i for i in data_raw.columns if i != target]
len(features_all)
```

The first thing we'll think about doing is making sure that our features are normalized.  After that, it could be helpful to do PCA analysis to see what 

```python
feature_data = data_raw[features_all].copy()
label_data = data_raw[target].copy()


#normalize feature_data, though not important with a Random Forest.
from sklearn import preprocessing
feat_normalized = preprocessing.normalize(feature_data, norm='l2')
print(feat_normalized[:1])
feature_data.loc[1]
```

Now we are going to train our prediction machine.  Since each sample stands alone, it would be easier to just split the data very 

```python
#we try with random Forest
from sklearn.ensemble.forest import RandomForestRegressor
from sklearn.model_selection import cross_val_score

#make model 
random_regressor = RandomForestRegressor(max_depth=None)
scores = cross_val_score(random_regressor, 
                         data_raw[features_all][1:].values, 
                         data_raw[target][1:].values, 
                         cv=5, 
                         scoring='neg_mean_squared_error')
print(scores[:3])
#example

print("Negative Mean Squared Error, signs are flipped.  It's a scikit learn thing. closer to 0 is better")
print("with two standard deviations: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))

forest_scores = cross_val_score(random_regressor, feat_normalized, label_data, cv=5, scoring='r2')
print("\n R2.  Closer to 1 is better")
print("with two standard deviations: %0.2f (+/- %0.2f)" % (forest_scores.mean(), forest_scores.std() * 2))
```

Let's graph the predictions against the actual labels.  

```python
#let's try with Keras, and an RNN
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import KFold
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

this_batch_size = 404 #for debugging
def neural_net():
# create model
	model = Sequential()
	model.add(Dense( 11, 
                     input_dim=11, 
                     init='normal', 
                     activation='relu',
                     batch_input_shape = (this_batch_size, 11))
              )
    
	model.add(Dense(22, init='normal', activation='relu', batch_input_shape = (this_batch_size, 22)))
	model.add(Dense(1, init='normal'))
	# Compile model
	model.compile(loss='mean_squared_error', optimizer='adam')
	return model


#because the pipeline does Standard Scaler normalization, we'll use non-normalized data
estimators = []
estimators.append(('standardize', StandardScaler()))
estimators.append(('mlp', KerasRegressor(
                            build_fn = neural_net, 
                            nb_epoch=50, 
                            batch_size = this_batch_size)))

pipeline = Pipeline(estimators)

results = cross_val_score(pipeline, 
                           feat_normalized, 
                           data_raw[target].values, 
                           cv=5, 
                           scoring = 'neg_mean_squared_error')

print("R2 score: %.2f (%.2f) R2" % (results.mean(), results.std()))

results = cross_val_score(pipeline, 
                           feat_normalized, 
                           data_raw[target].values, 
                           cv=5, 
                           scoring = 'r2')

print("R2 score: %.2f (%.2f) R2" % (results.mean(), results.std()))
```

This is a very simple, very easy implementation of machine learning on various indicators for housing price.  

The purpose of this code is to compare the performance of the different learning methods.  These will all be predicted value versus measured value results.

So far, what we have learned is that Random Forest is doing well, and a 2 layer Sequential neural network is doing really poorly.  

```python
#Sadly, we have to remake our models and re-test them because cross validation method doesn't
#keep our models in memory

#split 80:20 traning and testing. 
train_index = np.random.choice(range(0, data_raw.shape[0]), int(data_raw.shape[0] * 0.8))
test_index = [i for i in range(0, data_raw.shape[0]) if i not in train_index]

training = data_raw.ix[train_index][features_all]

training_features = preprocessing.normalize(data_raw.ix[train_index][features_all].as_matrix(), norm='l2')
training_labels = data_raw.ix[train_index][target]

testing_features = preprocessing.normalize(data_raw.ix[test_index][features_all].as_matrix(), norm='l2')
testing_labels = data_raw.ix[test_index][target]


#train both models
random_regressor.fit(training_features, training_labels)
my_neural = neural_net()
my_neural.fit(training_features, training_labels, nb_epoch=50)

#get predictions
random_predictions = random_regressor.predict(testing_features)
neural_predictions = my_neural.predict(testing_features)
```

```python
#Okay!  Now we are ready for graphing
import matplotlib.pyplot as plt

%matplotlib inline 

#tree
random_scatter, = plt.plot(testing_labels, random_predictions, 'o', color = 'red', label = 'random forest')
neural_scatter, = plt.plot(testing_labels, neural_predictions, 'o', color = 'blue', label = 'neural network')
plt.plot(range(0, 50), '-', color = 'black')

plt.title('Random Forest Regression Predicted Values versus Measured Values')
plt.legend(handles = [random_scatter, neural_scatter])
plt.xlabel('measured values by thousand USD')
plt.ylabel('predicted_values by thousand USD')
```
